# Zmiana typu danych

```python
# Sprawdzenie roku
print(autos['Year of registration'].unique())  # Zobacz, czy są dziwne wartości
print(autos['Year of registration'].describe())  # Zobacz min i max wartości
```

Wygląda na to że kolumna 'Year of registration' ma kilka dziwnych wartości jak 1111, 4800, 1001 czy 6200. W związku z tym należy się ich pozbyć aby przekonwertować na typ datetime.

```python
# Usuwamy nielogiczne wartości lat (np. <1900 i >2025)
autos = autos[(autos['Year of registration'] >= 1900) & (autos['Year of registration'] <= 2025)]
```

```python
# Sprawdzenie miesięcy
print(autos['Month of registration'].unique())  # Zobacz, czy są dziwne wartości
print(autos['Month of registration'].describe())  # Zobacz min i max wartości
```

```python
# Usuwamy nielogiczne wartości miesięcy
autos = autos[(autos['Month of registration'] >= 1) & (autos['Month of registration'] <= 12)]

# Utworzenie nowej kolumny z pełną datą rok-miesiąc
date = pd.to_datetime(autos['Year of registration'].astype(str) + '-' + autos['Month of registration'].astype(str) + '-01')
loc = autos.columns.get_loc('Year of registration')
autos.insert(loc, 'Registration Date', date)
del autos['Year of registration']
del autos['Month of registration']
```

```python
autos['Date Crawled'] = pd.to_datetime(autos['Date Crawled'], format='%Y-%m-%d %H:%M:%S')
autos['Price'] = autos['Price'].str.replace('$', '').str.replace(',', '').astype('float64')
autos['Unrepaired damage'] = autos['Unrepaired damage'].astype('bool')
autos['Ad created'] = pd.to_datetime(autos['Ad created'])
autos['Last Seen'] = pd.to_datetime(autos['Last Seen'], format='%Y-%m-%d %H:%M:%S')

autos.dtypes
```

```python
autos.rename({'Odometer': 'Odometer in km'}, axis=1, inplace=True)
autos['Odometer in km'] = autos['Odometer in km'].str.replace('km', '').str.replace(',', '').astype('float64')
autos['Odometer in km'].dtype
```

# Wartości odstające

```python
num_cols = autos.select_dtypes(include=np.number).columns
outlier_counts = {}

# Tworzymy kopię danych, aby nie nadpisywać oryginalnych wartości
autos_cleaned = autos.copy()

for col in num_cols:
    q1, q3 = autos[col].quantile([.25, .75])
    iqr = q3 - q1
    lower_bound, upper_bound = q1 - 1.5 * iqr, q3 + 1.5 * iqr
    outliers = autos[(autos[col] < lower_bound) | (autos[col] > upper_bound)]
    outlier_counts[col] = outliers.shape[0]
    autos_cleaned = autos_cleaned[(autos_cleaned[col] >= lower_bound) & (autos_cleaned[col] <= upper_bound)]

display(pd.Series(outlier_counts).to_frame("Liczba wartości odstających"))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Wybór kolumn do analizy
df_boxplot = autos[['Price', 'Power PS', 'Odometer in km']]

# Tworzenie subplots (3 wiersze, 1 kolumna)
fig, axes = plt.subplots(3, 1, figsize=(8, 12))  # 3 wiersze, 1 kolumna

columns = ['Price', 'Power PS', 'Odometer in km']

# Tworzenie boxplotów dla każdej kolumny z logarytmiczną skalą Y
for i, col in enumerate(columns):
    sns.boxplot(y=df_boxplot[col], ax=axes[i])
    axes[i].set_yscale('log')  # Skala logarytmiczna na osi Y
    axes[i].set_title(f'Boxplot dla {col} (Skala log)')
    axes[i].set_ylabel('Wartość (log scale)')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```
