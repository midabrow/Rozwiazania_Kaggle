### **_Project Goal:_**

Build and train a PyTorch-based MLP regression model that can predict the **market price of used cars** from a set of features such as brand, year, transmission, mileage, and fuel type.

This is a regression task involving structured tabular data, making it an ideal challenge to practice **feature preprocessing**, **deep learning with non-image data**, and **evaluation of regression performance**.

* * *

### **_Key Steps:_**

#### 1\. **Understand the problem**

-   Predict car prices based on available structured features.
    
-   The target is a **continuous variable** (regression).
    
-   The dataset contains **categorical** and **numerical** features.
    

#### 2\. **Data exploration (EDA)**

-   Analyze the distribution of the target variable (price).
    
-   Understand the value ranges and cardinality of each feature.
    
-   Identify potential outliers, missing values, skewness.
    

#### 3\. **Data preprocessing**

-   Fill or remove missing values.
    
-   Encode categorical features (e.g., Label Encoding or One-Hot).
    
-   Normalize numerical columns (e.g., mileage, engine size).
    
-   Prepare `X` and `y` for PyTorch model (train/test split, tensors).
    

#### 4\. **Build MLP model**

-   Define a **regression MLP** using PyTorch (`nn.Module`).
    
    -   Input layer: feature size
        
    -   Hidden layers: Linear + ReLU
        
    -   Output layer: 1 neuron (for continuous price)
        
-   Use `MSELoss` and `Adam` optimizer.
    

#### 5\. **Train and evaluate**

-   Train over several epochs and track loss.
    
-   Evaluate performance using:
    
    -   RMSE
        
    -   MAE
        
    -   R¬≤ Score
        
-   Visualize predictions vs true prices.
    

#### 6\. **Summary and improvements**

-   Analyze which features influence the model most.
    
-   Identify under/over-fitting.
    
-   Consider future steps:
    
    -   Hyperparameter tuning
        
    -   Model ensembles
        
    -   Other regression architectures

# <a id='toc10_'></a>[<b><span style='color:darkorange'>Step 7.1 |</span><span style='color:#1E3A8A'> Data Preparation</span></b>](#toc0_)

```python
input_cols = [col for col in df.select_dtypes(include='number').columns if col != 'selling_price']
categorical_cols = df.select_dtypes(include='object').columns
output_cols = ['selling_price']
```

```python
def dataframe_to_arrays(dataframe):
    dataframe_copy = df.copy(deep=True)
    
    for col in categorical_cols:
        dataframe_copy[col] = dataframe_copy[col].astype('category').cat.codes

    input_array = dataframe_copy[input_cols].to_numpy()
    target_array = dataframe_copy[output_cols].to_numpy()

    return input_array, target_array

input_array, target_array = dataframe_to_arrays(df)
input_array
```

```python
target_array
```

```python
input = torch.Tensor(input_array)
target = torch.Tensor(target_array)

dataset = TensorDataset(input, target)

# 80% and 20%
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size

train_ds, val_ds = random_split(dataset, [train_size, val_size])

batch_size = 128

train_loader = DataLoader(dataset, batch_size, shuffle=True)
val_loader = DataLoader(val_ds, batch_size)
```

<a id="creating-model"></a>
# <a id='toc10_'></a>[<b><span style='color:darkorange'>Step 7.2 |</span><span style='color:#1E3A8A'> Creating model</span></b>](#toc0_)

```python
input_size = len(input_cols)
output_size = len(output_cols)



class CarsModel(nn.Module):
    def __init__(self):
        super().__init__()
        # to `nn.Linear(6, 1)` = jedna warstwa, kt√≥ra uczy siƒô mapowaƒá cechy ‚Üí cena.
        self.linear = nn.Linear(input_size, output_size)

    # m√≥wi: ‚Äûjak dane przechodzƒÖ przez model‚Äù
    def forward(self, xb):
        out = self.linear(xb)
        return out
    
    # ‚Äûco siƒô dzieje podczas 1 kroku treningowego‚Äù.
    def training_step(self, batch):
        inputs, targets = batch
        # Robi predykcjƒô out = self(inputs)
        out = self(inputs)
        # Oblicza b≈ÇƒÖd (L1 = MAE ‚Äì Mean Absolute Error)
        loss = F.l1_loss(out, targets)
        # Zwraca stratƒô do backward() i optimizer.step()
        return loss
        
    def validation_step(self, batch):
        inputs, targets = batch
        out = self(inputs)
        loss = F.l1_loss(out, targets)
        # .detach() = od≈ÇƒÖcza tensor od grafu obliczeniowego (bez grad)
        return {'val_loss': loss.detach()}

    def validation_epoch_end(self, outputs):
        batch_losses = [x['val_loss'] for x in outputs]
        epoch_loss = torch.stack(batch_losses).mean()
        return {'val_loss': epoch_loss.item()}

    def epoch_end(self, epoch, result, num_epoch):
        # Drukuje tylko co 20 epok albo na ko≈Ñcu
        if (epoch+1) % 20 == 0 or epoch == num_epoch -1:
            print(f"Epoch [{epoch+1}], val_loss: {result['val_loss']}")
```

```python
model = CarsModel()

list(model.parameters())
```

<a id="creating-model"></a>
# <a id='toc10_'></a>[<b><span style='color:darkorange'>Step 7.3 |</span><span style='color:#1E3A8A'> Training model</span></b>](#toc0_)

```python
def evaluate(model, val_loader):
    outputs = [model.validation_step(batch) for batch in val_loader]
    return model.validation_epoch_end(outputs)

# Fitting algorithm
def fit(epochs, lr, model, train_loader, val_loader, opt_func=torch.optim.SGD):
    history = []
    optimizer = opt_func(model.parameters(), lr)
    for epoch in range(epochs):
        # Training Phase 
        for batch in train_loader:
            loss = model.training_step(batch)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
        # Validation phase
        result = evaluate(model, val_loader)
        model.epoch_end(epoch, result, epochs)
        history.append(result)
    return history

# Check the initial value that val_loss have
result = evaluate(model, val_loader)
print(result)
```

```python
# Start with the Fitting
epochs = 90
lr = 1e-8
history1 = fit(epochs, lr, model, train_loader, val_loader)
```
